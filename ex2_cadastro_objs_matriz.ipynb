```python
# 06/10/2023
# Brennon Meireles
# Senai Suiço Brasileiro
# Exercício de FPOO

#  2. Desenvolver um programa que leia e exiba os elementos para uma matriz com qualquer número de linhas e colunas do tipo char.

#  input para o recebimento dos valores para a matriz:
linhas = int(input("Digite o nº de Linhas: "))
colunas = int(input("Digite o nº de Colunas: "))
print()

matriz = [] # matriz principal

# preenchimento da matriz com input:
for i in range(linhas):
  linha = [] # lista feita para colocar os valores do for dentro da matriz principal
  for j in range(colunas):
    elemento = input(f'Digite o elemento na posição ({i}),({j}): ')
    linha.append(elemento) # adição dos elementos para a lista
  matriz.append(linha) # adição de todos os elementos do for para a matriz principal
  print() # pra pular linha...

# saida:
print("Matriz digitada:\n")
for linha in matriz:
  for elemento in linha:
    print(f'Elementos:  {elemento}')

```

```python
# versão resumida:

# Input para o recebimento dos valores para a matriz:
linhas = int(input("Digite o nº de Linhas: "))
colunas = int(input("Digite o nº de Colunas: "))
print()

# Preenchimento da matriz com input:
matriz = []  # Matriz principal

for i in range(linhas):
    linha = []  # Lista para colocar os valores do for dentro da matriz principal
    for j in range(colunas):
        elemento = input(f'Digite o elemento na posição ({i}),({j}): ')
        linha.append(elemento)  # Adição dos elementos para a lista
    matriz.append(linha)  # Adição de todos os elementos do for para a matriz principal
    print()  # Para pular linha...

# Saída:
print("Matriz digitada:\n")
for linha in matriz:
    print(" ".join(linha))  # Usando join para imprimir a linha como uma string separada por espaço
```
